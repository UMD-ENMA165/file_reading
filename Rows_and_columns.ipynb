### list-of-lists function

```python
def get_list_of_lists(fname):
    data_file = open(fname)
    data = data_file.readlines()
    data_file.close()

    list_of_lists = []
    for line in data:
        strip = line.rstrip()
        split = strip.split(",")
        list_of_ints = []
        for string in split:
            number = int(string)
            list_of_ints.append(number)
        list_of_lists.append(list_of_ints)
    
    return list_of_lists
```

```python
lol = get_list_of_lists("data/inflammation-03.csv")
```

```python
def maximum_of_columns(list_of_lists):
    max_columns = []
    c = 0
    r = 0
    while c < len(list_of_lists[0]):
        m = -1
        r = 0
        while r < len(list_of_lists):
            number = list_of_lists[r][c]
            if number > m:
                m = number

            r = r+1

        max_columns.append(m)

        c = c+1

    return max_columns
```

```python
def maximum_of_rows(list_of_lists):
    sumlist = []
    r = 0 
    while r < len(list_of_lists):
        m = -float("inf")
        c = 0
        while c < len(list_of_lists[r]):
            number = list_of_lists[r][c]
            if number > m:
                m = number

            c = c+1

        sumlist.append(m)

        r = r+1
    return sumlist
```

```python
def minimum_of_rows(list_of_lists):
    sumlist = []
    r = 0 
    while r < len(list_of_lists):
        m = float("inf")
        c = 0
        while c < len(list_of_lists[r]):
            number = list_of_lists[r][c]
            if number < m:
                m = number

            c = c+1

        sumlist.append(m)

        r = r+1
    return sumlist
```

```python
def comparison_of_rows(list_of_lists, ismax):
    sumlist = []
    r = 0 
    while r < len(list_of_lists):
        if ismax == True:
            m = -float("inf")
        else:
            m = float("inf")
        c = 0
        while c < len(list_of_lists[r]):
            number = list_of_lists[r][c]
            if ismax == True:
                if number > m:
                    m = number
            else:
                if number < m:
                    m = number

            c = c+1

        sumlist.append(m)

        r = r+1
    return sumlist
```

```python
def comparison_of_rows2(list_of_lists, initial_value, compare):
    sumlist = []
    r = 0 
    while r < len(list_of_lists):
        m = initial_value
        c = 0
        while c < len(list_of_lists[r]):
            number = list_of_lists[r][c]
            if compare(number, m):
                m = number

            c = c+1

        sumlist.append(m)

        r = r+1
    return sumlist
```

```python
def max_of_rows(list_of_lists):
    return comparison_of_rows(list_of_lists, ismax=True)
```

```python
def min_of_rows(list_of_lists):
    return comparison_of_rows(list_of_lists, ismax=False)
```

```python
def greater(a, b):
    return a > b
```

```python
def lesser(a, b):
    return a < b
```

```python
def max_of_rows2(list_of_lists):
    return comparison_of_rows2(list_of_lists, 
                               initial_value=-float("inf"),
                               compare=greater)
```
