## List of list of integers

```python
list_of_lists = []
for line in data:
    strip = line.rstrip()
    split = strip.split(",")
    list_of_ints = []
    for string in split:
        number = int(string)
        list_of_ints.append(number)
    list_of_lists.append(list_of_ints)
    
print(list_of_lists)
```
